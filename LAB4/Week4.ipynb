```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
nltk.download("punkt_tab")
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text1="Twas brillig, and the slithy toves Did gyre and gimble in the wabe: All mimsy were the borogoves"
text2= "And the mome raths outgrabe. Beware the Jabberwock, my son! The jaws that bite, the claws that catch!"
text3= "Beware the Jubjub bird, and shun The frumious Bandersnatch!"
```

```python
text = text1 + text2 + text3
words = word_tokenize(text)

pos_tags = pos_tag(words)
print("Original Text: ")
print(text)

print("\nPoS Tagging Results:")
for word, pos_tag in pos_tags:
  print(f"{word}{pos_tag}")
```

```python
import spacy
# Download the model

nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
```

```python
print("Original Text: ", text)
print("PoS Tagging Result:")
for token in doc:
    print(f"{token.text}: {token.pos_}")
```
